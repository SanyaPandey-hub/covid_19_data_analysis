```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
```

```python
sample_df = pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
sample_df.head()
```

```python
sample_df = sample_df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
sample_df.columns = ['date', 'state','cured','deaths','confirmed']
```

```python
sample_df.head()
```

```python
sample_df.head(100)
```

```python
sample_df.dropna()
```

```python
today = sample_df[sample_df.date=='2021-01-03']
```

```python
today.head()
```

```python
today.shape
```

```python
today.head()
```

```python
max_confirmed_cases=today.sort_values(by='confirmed',ascending=False)
```

```python
max_confirmed_cases.head()
```

```python
top_states_confirmed = max_confirmed_cases[0:10]
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="state",y="confirmed",data=top_states_confirmed,hue="state")
plt.show()
```

```python
max_death_cases=today.sort_values(by='deaths',ascending=False)
```

```python
max_death_cases.head()
```

```python
top_states_death = max_death_cases[0:5]
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="state",y="deaths",data=top_states_confirmed,hue="state")
plt.show()
```

```python
max_cured_cases=today.sort_values(by='deaths',ascending=False)
```

```python
max_cured_cases.head()
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="state",y="cured",data=top_states_confirmed,hue="state")
plt.show()
```
